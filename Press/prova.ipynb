```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
rho = np.fromfile("Al5_3/gasdens5.dat").reshape(384,1152)

# Creare un grafico in coordinate polari
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')

# Imposto raggio max scala
ax.set_rmax(4.0/12.0)
ax.set_rmin(0)

#Creo griglia polare
theta = np.linspace(-np.pi, np.pi, 1152)
r = np.linspace(0.4/12.0, 4.0/12.0, 384)

# Creare una griglia di valori R e Theta
Theta, R = np.meshgrid(theta, r)

cax = plt.pcolormesh(Theta, R, np.log10(rho), cmap='jet')
plt.colorbar(cax, label='Densit')
plt.title("Disco: setup fargo")

#Imposto scala raggi
scala_raggi_desiderati = [0.07, 0.14, 0.21, 0.28]
ax.set_yticks(scala_raggi_desiderati)

# Mostrare il grafico
plt.show()
```
